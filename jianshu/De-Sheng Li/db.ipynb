```python
import sqlite3
import pandas as pd
```

```python
def commit_sql(db, sql):
    """
        将所做的修改保存到数据库
        
        @param db: database name, string
        @param sql: sql statement, string
    """
    conn = sqlite3.connect(db + '.db')
    
    cursor = conn.cursor() #创建 Cursor
    cursor.execute(sql)
    cursor.close()
    
    conn.commit()
    conn.close()
    
    print("提交成功")
```

```python
# 创建表
# 表名：mzdxj
# 字段：id, title, date, vol, period, intro, content, anno
create_mao = """
    CREATE TABLE mzdxj (
        id INTEGER PRIMARY KEY AUTOINCREMENT, --自增主键
        title VARCHAR(255), --文章标题
        date TEXT, --日期，格式YYYY-MM-DD
        vol INTEGER, --卷次
        period VARCHAR(40), --时期
        intro TEXT, --文章介绍
        content TEXT, --文章内容
        anno TEXT --文章注释
    )"""

commit_sql('mao', create_mao)
```

```python
def insert_into_mzdxj():
    """
        插入一条记录到选集表
    """
    title = input("文章标题：")
    date = input("日期（YYYY-MM-DD）：")
    vol = input("卷次：")
    period = input("时期：")
    intro = input("文章介绍：")
    content = input("文章内容：")
    anno = input("文章注释：")
    
    insert_sql = """
    INSERT INTO mzdxj (id, title, date, vol, period, intro, content, anno) 
        VALUES(
            NULL,
            '%s',
            '%s',
            %d, 
            '%s',
            '%s',
            '%s',
            '%s'
        )
    """ % (title, date, int(vol), period, intro.replace(" ", ""), content.replace(" ", ""), anno.replace(" ", ""))
    
    commit_sql('mao', insert_sql)
```

```python
df = pd.read_sql('select * from mzdxj', con=sqlite3.connect('mao.db'))
print("Record count: {}".format(df['title'].count()))
df.tail()
```

```python
n = 4

for i in range(n):
    print(i)
    insert_into_mzdxj()
    print("\n")
```
