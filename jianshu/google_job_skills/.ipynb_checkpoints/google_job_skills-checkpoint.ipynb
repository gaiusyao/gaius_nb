```python
# 科学计算
import numpy as np
# 数据处理及导入导出
import pandas as pd

# 数据可视化
import matplotlib.pyplot as plt
plt.style.use('ggplot') #使用 ggplot 主题
# 更好的可视化效果
import seaborn as sns
sns.set_style("whitegrid") #设置 seaborn 主题

# 词云
from wordcloud import WordCloud, STOPWORDS

# 正则表达式
import re
# 默认字典
from collections import defaultdict
```

```python
# 导入数据集
df = pd.read_csv('job_skills.csv')
```

```python
df.info()
```

```python
df.sample(5)
```

```python
# 修改 DataFrame 列名
df.columns = ['公司名称', '职位名称', '职位类别', '工作地点', '岗位职责', '最低资格', '首选资格']
```

```python
# 检查是否含有 NaN 值
pd.isnull(df).sum()
```

```python
# 删除含 NaN 值的列
df = df.dropna(how='any', axis='rows')
```

```python
df['公司名称'].value_counts()
```

```python
# 删除 YouTube
df = df[df['公司名称'] != 'YouTube']
```

```python
df['工作城市'] = df['工作地点'].apply(lambda x : x.split(',')[0])
df['工作国家'] = df['工作地点'].apply(lambda x : x.split(',')[-1])
```

```python
def top15_bar(col):
    """
        生成前 15 的柱状图
        
        Args:
            col: column name, string
    """
    f, ax = plt.subplots(figsize=(12, 6))
    
    df_top15 = df[col].value_counts()[0:15]
    
    x = df_top15.values
    y = df_top15.index
    
    sns.set_color_codes("muted")
    sns.barplot(x=x, y=y, color='b');
```

```python
top15_bar('工作国家')
```

```python
top15_bar('工作城市')
```

```python
top15_bar('职位类别')
```
