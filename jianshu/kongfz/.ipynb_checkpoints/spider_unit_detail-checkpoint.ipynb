```python
url = 'http://item.kongfz.com/book/29448359.html'

my_headers = {
    'Host': 'item.kongfz.com',
    'Referer': 'http://search.kongfz.com/item_result/?select=0&key=%E8%B5%84%E6%9C%AC%E8%AE%BA&author=hk9a6ck514bk601d&order=2',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.81 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest',
}

res = requests.post(url=url, headers = my_headers)  
res.raise_for_status()  
res.encoding = 'utf-8'  
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.title.get_text()
```

```python
soup.link['href']
```

```python
info =""

for item in soup.find('div', 'detail-con-right-top').find_all('div', 'item'):
    text_key = item.find('span', 'text-key').get_text()
    text_value = item.find('span', 'text-value').get_text()
    info += text_key + text_value + '\n'
    
print(info)
```

```python
intro = ""

for li in soup.find_all('li', 'jianjie'):
    title = li.h5.extract().get_text()
    tmp = li.prettify()
    pattern = re.compile('<(.|\n)+?>\\n\s*')
    content = re.sub(pattern, '', tmp)
    intro += title + "\n" + content
    
print(intro)
```

```python
book_count = int(soup.find('li', 'gray3 p_l20 fontBlod').get_text()[:-1])
page_num = math.ceil(book_count / 50)
```

```python
top50_product_list = []

for product in soup.find('div', 'detail-list-con').find_all('li', 'item-list'):
    product_info = []
    
    product_info.append(product['itemid'])
    product_info.append(product['shopid'])
    product_info.append(product.find('div', 'list-con-title').find('a').get_text())
    product_info.append(product.find('div', 'list-con-title').find('a')['href'])
    product_info.append(product.find('div', 'list-con-title').find('span', 'list-con-dis').get_text())
    product_info.append(product.find('div', 'list-con-product').get_text())
    product_info.append(product.find('div', 'list-con-press').find('div', 'con-press-title').a.get_text().strip())
    product_info.append(product.find('div', 'list-con-press').find('div', 'con-press-title').a['href'])
    product_info.append(product.find('div', 'list-con-press').find('div', 'con-press-city').get_text())
    product_info.append(product.find('div', 'list-con-press').find('a', rel='nofollow')['class'][0])
    product_info.append(float(product.find('div', 'list-con-moneys').span.get_text())) 
    
    top50_product_list.append(product_info)
    
df = pd.DataFrame(data = top50_product_list, columns = ['itemid', 'shopid', 'title', 'detail_url', 'subtitle', 'quality', 'shop', 'shop_url', 'shop_city', 'shop_star', 'prize'])
```
